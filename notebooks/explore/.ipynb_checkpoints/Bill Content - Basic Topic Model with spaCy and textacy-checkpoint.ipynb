## Textual Analysis

### Preproc and Build Corpus

```python
textacy_preprocessor = lambda text: textacy.preprocess.preprocess_text(text,
                                                                       no_contractions=True,
                                                                       no_numbers=True,
                                                                       no_emails=True,
                                                                       no_currency_symbols=True,
                                                                       lowercase=True)
```

```python
nlp = spacy.load("en", add_vectors=False)
```

```python
nlp.pipeline = [nlp.tagger, nlp.parser]
```

```python
# "texts"
texts = [textacy_preprocessor(con) for con in list(btfc.content)]
```

```python
metas = [btfc.ix[btfc.index[i]][["session", "house", "bill", "keywords"]].to_dict() for i in range(btfc.shape[0])]
```

```python
corp = textacy.Corpus(lang=nlp)
```

```python
corp.add_texts(texts=texts, metadatas=metas,
               n_threads=6, batch_size=250)
```

```python
corp.n_docs
```

```python
corp.n_sents
```

### Topic Model

#### Tokenize and Vectorize the corpus:

```python
terms_lists = (doc.to_terms_list(ngrams=1, named_entities=True, as_strings=True) for doc in corp)
doc_term_matrix, id2term = textacy.vsm.doc_term_matrix(\
                                                       terms_lists, 
                                                       weighting='tfidf', normalize=True, smooth_idf=True,
                                                       min_df=3, max_df=0.95, max_n_terms=100000)
doc_term_matrix
```

#### Init and Train a Topic Model:

```python
model = textacy.tm.TopicModel('nmf', n_topics=20)
model.fit(doc_term_matrix)
model
```

#### Transform the corpus and Interpret the Model:

```python
doc_topic_matrix = model.transform(doc_term_matrix)
```

```python
for topic_idx, top_terms in model.top_topic_terms(id2term, topics=range(5)):
    print('topic', topic_idx, ':', '   '.join(top_terms))
```

```python
for topic_idx, top_docs in model.top_topic_docs(doc_topic_matrix, topics=[0,1], top_n=2):
    print(topic_idx)
    for j in top_docs:
        print(corp[j].metadata['keywords'])
```

```python
for i, val in enumerate(model.topic_weights(doc_topic_matrix)):
    print(i, val)
```

#### Visualize the Model

```python
model.termite_plot(doc_term_matrix, id2term,
                   topics=-1,  n_terms=50, sort_terms_by='seriation');
```
